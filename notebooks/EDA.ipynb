```python
import pandas as pd
import streamlit as st
import plotly.express as px
import altair as at

df_cars = pd.read_csv('https://github.com/DC89/sprint4pjkt/blob/main/vehicles_us.csv')
                
df_cars.info()
```

```python
df_cars.head(25)
```

```python
df_cars['is_4wd'] = df_cars['is_4wd'].fillna(0)
```

```python
df_cars.head(10)
```

```python
def OverOneHundredThousand(df_cars):
    df_cars['100k'] = df_cars['odometer'].apply(lambda x: 1 if x >= 100000.0 else 0)
    return df_cars
   
```

```python
OverOneHundredThousand(df_cars)
```

```python
df_cars['model_year'] = df_cars['model_year'].fillna(0)
df_cars.head()
```

```python
fig1 = px.scatter(x=df_cars['odometer'], y=df_cars['days_listed'])
fig1.show()
```

```python
df_carsNew = df_cars.drop(df_cars[df_cars['model_year'] <= 1999.0].index)
fig2 = px.histogram(df_carsNew, x='model_year')
fig2.show()
```

```python
fig3 = px.scatter(x=df_carsNew['model_year'], y=df_carsNew['days_listed'])
fig3.show()
```

```python
df_carsClassic = df_cars.drop(df_cars[(df_cars['model_year'] > 1999.0) | (df_cars['model_year'] == 0)].index)
```

```python
fig4 = px.histogram(df_carsClassic, x='model_year')
fig4.show()
```

```python
fig5 = px.scatter(x=df_carsClassic['model_year'], y=df_carsClassic['days_listed'])
fig5.show()
```
